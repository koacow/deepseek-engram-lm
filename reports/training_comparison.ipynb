```python
import pandas as pd
import matplotlib.pyplot as plt

baseline_log = pd.read_csv('../checkpoints/baseline_log.csv')
engram_log = pd.read_csv('../checkpoints/engram_log.csv')
engram_frozen_log = pd.read_csv('../checkpoints/engram_frozen_log.csv')
params_log = pd.read_csv('../checkpoints/params_log.csv')
```

```python
# Plot the training loss for baseline and engram models
plt.figure(figsize=(10, 6))
plt.plot(baseline_log['step'], baseline_log['train_loss'], label='Baseline', color='blue')
plt.plot(params_log['step'], params_log['train_loss'], label='Parameter Control', color='red')
plt.plot(engram_log['step'], engram_log['train_loss'], label='Engram', color='orange')
plt.plot(engram_frozen_log['step'], engram_frozen_log['train_loss'], label='Engram w/ Frozen Gate', color='green')
plt.plot()
plt.ylim(2, 5)
plt.xlabel('Training Steps')
plt.ylabel('Training Loss')
plt.title('Training Loss Comparison')
plt.legend()
plt.grid()
plt.savefig('./figs/training_loss_comparison.png')
plt.show()
```

```python
print("Lowest training loss for each model:")
print(f"Baseline: {baseline_log['train_loss'].min():.4f} at step {baseline_log['step'][baseline_log['train_loss'].idxmin()]}")
print(f"Parameter Control: {params_log['train_loss'].min():.4f} at step {params_log['step'][params_log['train_loss'].idxmin()]}")
print(f"Engram: {engram_log['train_loss'].min():.4f} at step {engram_log['step'][engram_log['train_loss'].idxmin()]}")
print(f"Engram w/ Frozen Gate: {engram_frozen_log['train_loss'].min():.4f} at step {engram_frozen_log['step'][engram_frozen_log['train_loss'].idxmin()]}")
print("=" * 10)
print("Number of training steps for each model:")
print(f"Baseline: {baseline_log['step'].iloc[-1]}")
print(f"Parameter Control: {params_log['step'].iloc[-1]}")
print(f"Engram: {engram_log['step'].iloc[-1]}")
print(f"Engram w/ Frozen Gate: {engram_frozen_log['step'].iloc[-1]}")
print("=" * 10)
print("Final training loss for each model:")
print(f"Baseline: {baseline_log['train_loss'].iloc[-1]:.4f}")
print(f"Parameter Control: {params_log['train_loss'].iloc[-1]:.4f}")
print(f"Engram: {engram_log['train_loss'].iloc[-1]:.4f}")
print(f"Engram w/ Frozen Gate: {engram_frozen_log['train_loss'].iloc[-1]:.4f}")
```

```python
# Plot the validation perplexity for baseline and engram models
plt.figure(figsize=(10, 6))
plt.scatter(baseline_log['step'], baseline_log['val_ppl'], label='Baseline', color='blue')
plt.scatter(params_log['step'], params_log['val_ppl'], label='Parameter Control', color='red')
plt.scatter(engram_log['step'], engram_log['val_ppl'], label='Engram', color='orange')
plt.scatter(engram_frozen_log['step'], engram_frozen_log['val_ppl'], label='Engram w/ Frozen Gate', color='green')
plt.ylim(0, 100)
plt.xlabel('Training Steps')
plt.ylabel('Validation Perplexity')
plt.title('Validation Perplexity Comparison')
plt.legend()
plt.grid()
plt.savefig('./figs/validation_perplexity_comparison.png')
plt.show()
```

```python
print("Lowest validation perplexity for each model:")
print(f"Baseline: {baseline_log['val_ppl'].min():.4f} at step {baseline_log['step'][baseline_log['val_ppl'].idxmin()]}")
print(f"Parameter Control: {params_log['val_ppl'].min():.4f} at step {params_log['step'][params_log['val_ppl'].idxmin()]}")
print(f"Engram: {engram_log['val_ppl'].min():.4f} at step {engram_log['step'][engram_log['val_ppl'].idxmin()]}")
print(f"Engram w/ Frozen Gate: {engram_frozen_log['val_ppl'].min():.4f} at step {engram_frozen_log['step'][engram_frozen_log['val_ppl'].idxmin()]}")
print("=" * 10)
print("Final validation perplexity for each model:")
print(f"Baseline: {baseline_log['val_ppl'].dropna().iloc[-1]:.4f}")
print(f"Parameter Control: {params_log['val_ppl'].dropna().iloc[-1]:.4f}")
print(f"Engram: {engram_log['val_ppl'].dropna().iloc[-1]:.4f}")
print(f"Engram w/ Frozen Gate: {engram_frozen_log['val_ppl'].dropna().iloc[-1]:.4f}")
```
